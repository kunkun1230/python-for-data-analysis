### Counting time zones in pure Python

```python
#由于某些rec列表中没有tz，故出错了
time_zones = [rec['tz'] for rec in records]
```

```python
time_zones = [rec['tz'] for rec in records if 'tz' in rec]
```

```python
#统计列表中各个元素的出现次数
def get_counts(sequence):
    counts = {}
    for x in sequence:
        if x in counts:
            counts[x] += 1
        else:
            counts[x] = 1
    return counts
```

```python
from collections import defaultdict

def get_counts2(sequence):
    #生成一个字典，所有值初始化为0
    counts =defaultdict(int)
    for x in sequence:
        counts[x]+=1
    return counts
```

```python
counts = get_counts(time_zones)
```

```python
def top_counts(count_dict, n=10):
    #枚举字典中的数组，并形成列表
    value_key_pairs = [(count, tz) for tz, count in count_dict.items()]
    value_key_pairs.sort()
    return value_key_pairs[-n:]
```

```python
top_counts(counts)
```

```python
import pandas as pd
names1880 = pd.read_csv('ch02/names/yob1880.txt', names=['name', 'sex', 'births'])
names1880
```

```python
#按性别计算1880这段时间出生的男孩和女孩总数
names1880.groupby('sex').births.sum()
```

```python
# 2010年是目前最后一个有效的统计年度
years = range(1880, 2011)

pieces = []
columns = ['name', 'sex', 'births']

for year in years:
    path = 'ch02/names/yob%d.txt' % year
    frame = pd.read_csv(path, names=columns)

    frame['year'] = year
    pieces.append(frame)
```

```python
#将pieces中的各个行组合成一个较大的Dataframe（names），并不保留文件中所返回的原始行号
names = pd.concat(pieces, ignore_index=True)

# Concatenate everything into a single DataFrame
#names = pd.concat(pieces)
```

```python
names
```

```python
#以年为索引，统计各个年份男女的出生数量
total_births = names.pivot_table('births', index='year',
                                 columns='sex', aggfunc=sum)
```

```python
total_births.tail()
```

```python
total_births.plot(title='Total births by sex and year')
```

```python
def add_prop(group):
    # Integer division floors
    births = group.births.astype(float)

    group['prop'] = births / births.sum()
    return group
names = names.groupby(['year', 'sex']).apply(add_prop)
```

```python
#对结果的有效性进行检查，验证所有分组的prop总和是否为1
np.allclose(names.groupby(['year', 'sex']).prop.sum(), 1)
```

```python

def get_top1000(group):
    return group.sort_index(by='births', ascending=False)[:1000]
grouped = names.groupby(['year', 'sex']) #按照年和性别对数据进行归类
top1000 = grouped.apply(get_top1000)  #筛选出每年每个性别前1000名的数据，注意理解，并不是总的前1000数据
```

get_top1000函数与下面的for循环意义一样

```python
pieces = []
for year, group in names.groupby(['year', 'sex']):
    pieces.append(group.sort_index(by='births', ascending=False)[:1000])
top1000 = pd.concat(pieces, ignore_index=True)
```

```python
top1000.index = np.arange(len(top1000))
```

```python
top1000
```

```python
def get_quantile_count(group, q=0.5):
    group = group.sort_index(by='prop', ascending=False)
    return group.prop.cumsum().values.searchsorted(q) + 1

diversity = top1000.groupby(['year', 'sex']).apply(get_quantile_count)
```

#### The "Last letter" Revolution

```python
# extract last letter from name column
get_last_letter = lambda x: x[-1]
last_letters = names.name.map(get_last_letter)
last_letters.name = 'last_letter'

table = names.pivot_table('births', index=last_letters,
                          columns=['sex', 'year'], aggfunc=sum)
```

```python
subtable = table.reindex(columns=[1910, 1960, 2010], level='year')
subtable.head()
```

```python
subtable.sum()
```

```python
letter_prop = subtable / subtable.sum().astype(float)
letter_prop.head()
```

```python
import matplotlib.pyplot as plt

fig, axes = plt.subplots(2, 1, figsize=(10, 8))
letter_prop['M'].plot(kind='bar', rot=0, ax=axes[0], title='Male')
letter_prop['F'].plot(kind='bar', rot=0, ax=axes[1], title='Female',
                      legend=False)
```

#### Boy names that became girl names (and vice versa)

```python
#查找以lesl开头的名字在男女中的变化情况
all_names = top1000.name.unique()
mask = np.array(['lesl' in x.lower() for x in all_names])
lesley_like = all_names[mask]
lesley_like
```

```python
filtered = top1000[top1000.name.isin(lesley_like)]
filtered.groupby('name').births.sum()
```

```python
table = filtered.pivot_table('births', index='year',
                             columns='sex', aggfunc='sum')
table = table.div(table.sum(1), axis=0)
table.tail()
```
